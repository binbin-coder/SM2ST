```r
library(Cardinal)
path = '/home/llx/project/MSI_data/2025_2_21_mouse_brain/30um'
input_file = '/home/llx/project/MSI_data/2025_2_21_mouse_brain/30um/ITO-PET-glass-30um.rds'
output_name = 'ITO-PET-glass-30um_norm'
output_csv = 'ITO-PET-glass-30um_norm.csv'
tiny <- readRDS(file = input_file)#mouse_ITO_50um/
writeMSIData(tiny, file = paste(path, output_name, sep = "/"),
             folder = path, outformat = "imzML",
             mz.type="32-bit float", intensity.type="32-bit float")
numeric_list <- vector("character", length = length(coord(tiny)$x))
for (i in seq_along(coord(tiny)$x)) {
    numeric_list[i] <- paste0("x", coord(tiny)$x[i], "_y", coord(tiny)$y[i])
}
table<-spectra(tiny, "intensity")
colnames(table)<-numeric_list
rownames(table)<-mz(tiny)
write.csv(table, file = paste(path, output_csv, sep = "/"), row.names = TRUE)
```
